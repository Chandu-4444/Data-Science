```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../datasets/employee.csv')
```

```python
data
```

```python
# Fill missing values with mean
data['age'] = data['age'].fillna(data.age.mean())
data
```

```python
# Fill missing values in income column with median
data['income'] = data['income'].fillna(data.income.median())
data
```

```python
# Fill missing gender values with mode
data['gender']=data['gender'].fillna(data['gender'].mode()[0])
data
```

```python
# Handling outliers
```

```python
# Dropping outliers using standard deviation and mean
upper_limit = data['performance_score'].mean() + 2*data['performance_score'].std()
lower_limit = data['performance_score'].mean() - 2*data['performance_score'].std()

new_data = data[(data['performance_score'] < upper_limit) & (data['performance_score'] > lower_limit)]
new_data
```

```python
# Dropping outliers using Percentiles
upper_limit = data['performance_score'].quantile(.99)
lower_limit = data['performance_score'].quantile(.01)
new_data = data[(data['performance_score'] < upper_limit) & (data['performance_score'] > lower_limit)]
new_data
```

```python
# Feature Encoding Techniques
```

```python
# one-hot encoding
encoded_data = pd.get_dummies(data['gender'])
encoded_data = data.join(encoded_data)
encoded_data
```

```python
# One-hot encoding
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()

onehotencoder.fit_transform(data[['gender']]).toarray()
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

encoded_data = label_encoder.fit_transform(data['department'])

encoded_data
```

```python
# Inverting labels
decode_label = label_encoder.inverse_transform(encoded_data)
decode_label
```
