```python
data['gender'].replace(['female', 'male'], [0,1], inplace=True)
```

```python
data['gender'].fillna((data['gender'].mode()[0]), inplace=True)
```

```python
sns.countplot(data = data, x = 'gender')
```

```python
# Which category gender has more friends?
print('Friend count for male: ',data[data['gender']==1]['friend_count'].sum())
print('Friend count for female: ',data[data['gender']==0]['friend_count'].sum())
```

```python
# Which category gender has initiated more friends?
print('Friend count for male: ',data[data['gender']==1]['friendships_initiated'].sum())
print('Friend count for female: ',data[data['gender']==0]['friendships_initiated'].sum())
```
