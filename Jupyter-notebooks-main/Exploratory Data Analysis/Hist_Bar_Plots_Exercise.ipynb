```python
# data_2016_winners = data_2016.loc[(data_2016['Medal']=='Gold') | (data_2016['Medal']=='Silver') | (data_2016['Medal']=='Bronze')]
data_2016_winners = data_2016.loc[data_2016['Medal'].notnull()]
```

```python
data_2016_winners.shape
```

```python
data_2016_winners.head()
```

```python
top_sports = data_2016_winners.groupby(['Sport'], as_index=False).count().sort_values(by='Medal', ascending=False)[:5]
top_sports
```

```python
sns.barplot(x=top_sports['Sport'], y=top_sports['Medal'])
```

```python
top_performers = data_2016_winners[data_2016_winners['Sport'].isin(top_sports['Sport'])]
```

```python
top_performers['Sport'].unique()
```

```python
sns.histplot(top_performers['Age'])
```

```python
sns.catplot(x="Team", data=top_performers, kind="count", aspect=3).set_xticklabels(rotation=90)
```

```python
sns.catplot(x='Sport',y='Weight', data=top_performers, hue='Sex', kind='bar', estimator=np.mean, aspect=3)
```
