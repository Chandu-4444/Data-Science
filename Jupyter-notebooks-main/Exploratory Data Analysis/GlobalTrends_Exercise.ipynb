```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
olympics = pd.read_csv('../datasets/athlete_events.csv')
```

```python
# Task 1 - Plot weight vs height for 2016 winners of top 5 sports

# Filtering all 2016 sports
olympics_2016 = olympics[olympics['Year']==2016]

# Filtering all winners
olympics_2016_winners = olympics_2016.loc[olympics_2016['Medal'].notnull()]

# Filtering top 5 sports winners

## Filtering top 5 sports
top_sports = olympics_2016_winners.groupby(["Sport"], as_index=False).count().sort_values(by='Medal', ascending=False)[:5]
## Filtering top 5 sports winners
top_sports_winners = olympics_2016_winners[olympics_2016_winners["Sport"].isin(top_sports["Sport"])]

top_sports_winners.head()
```

```python
# Plotting weight vs height
sns.scatterplot(x="Weight", y="Height", hue="Sex", data=top_sports_winners)
```

```python
sns.jointplot(x="Weight", y="Height", kind="hex", data=top_sports_winners)
```

```python
sns.kdeplot(x="Weight", y="Height", data=top_sports_winners, shade=True, hue="Sex")
```

```python
sns.violinplot(x='Medal', y="Weight", hue='Sex', data=top_sports_winners)
```

```python
sns.violinplot(x="Medal", y="Height", hue="Sex", data=top_sports_winners)
```
