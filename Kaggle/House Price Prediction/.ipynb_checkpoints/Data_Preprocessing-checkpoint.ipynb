```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer

# plt.rcParams["figure.dpi"] = 600
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
print(train_df.shape)
train_df.head()
```

```python
print(test_df.shape)
test_df.head()
```

```python
joined_df = pd.concat([train_df.drop("SalePrice", axis=1), test_df])

print(joined_df.shape)
joined_df.head()
```

```python
def plot_missing(data, title="Missing values"):
    plt.figure(figsize=(28,10))
    sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
    plt.tick_params(labeltop=True, labelright=True)
    plt.xticks(rotation=90, fontsize=12)
    plt.title(title)
    plt.vlines(x=np.arange(0, data.shape[1]), ymin=0, ymax=data.shape[0], color='black', linestyle='--')
    plt.show()
```

```python

missing_percent = (joined_df.isnull().sum().sort_values(ascending=False) / joined_df.shape[0]) * 100

print(missing_percent[:10])
features_to_drop = missing_percent.index[:6]
print(features_to_drop.values)
```

```python
joined_df = joined_df.drop(columns=features_to_drop, axis=1)

plot_missing(joined_df)
```

```python
plt.figure(figsize=(28,10))
sns.heatmap(train_df.corr(), annot=True, cmap='viridis', linewidths=0.1)
```

### Filling Missing values with reasonable data

```python
# Utilities has only one class 1 variable. We'll remove this feature
joined_df = joined_df.drop("Utilities", axis=1)
```

```python
categorical_features = []
numerical_features = []

for column in joined_df.columns:
    if(joined_df[column].dtype == 'object' and joined_df[column].isna().any()):
        categorical_features.append(column)
    elif(joined_df[column].isna().any() and joined_df[column].dtype != 'object'):
        numerical_features.append(column)
```

```python
# categorical_features, numerical_features
```

```python
# print(joined_df["MasVnrType"].value_counts())

# imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
# joined_df["MasVnrType"] = imputer.fit_transform(np.array(joined_df["MasVnrType"]).reshape(-1, 1))
# plot_missing(joined_df)


for column in categorical_features:
    imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    joined_df[column] = imputer.fit_transform(np.array(joined_df[column]).reshape(-1, 1))
plot_missing(joined_df)
```

```python
for column in numerical_features:
    imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
    joined_df[column] = imputer.fit_transform(np.array(joined_df[column]).reshape(-1, 1))
plot_missing(joined_df)
```

```python
joined_df = pd.get_dummies(joined_df)
```

```python
train_df_preprocessed = joined_df.iloc[:len(train_df)]
test_df_preprocessed = joined_df.iloc[len(train_df):]
```

### Dealing with Outliers

```python
train_df_preprocessed = pd.concat([train_df_preprocessed, train_df['SalePrice']], axis=1)
```

```python
cols = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']

sns.pairplot(pd.concat([train_df_preprocessed[cols], train_df['SalePrice']], axis=1))
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(novelty=False)
preds = lof.fit_predict(np.array(train_df_preprocessed["GrLivArea"]).reshape(-1, 1))
```

```python
indexes = np.where(preds==-1)

indexes[0]
```

```python
sns.scatterplot(x = train_df_preprocessed['GrLivArea'], y = train_df['SalePrice'])
```

```python
train_df_preprocessed = train_df_preprocessed.drop(labels=indexes[0], axis=0)
```

```python
lof = LocalOutlierFactor(novelty=False)
preds = lof.fit_predict(np.array(train_df_preprocessed["TotalBsmtSF"]).reshape(-1, 1))
```

```python
indexes = np.where(preds==-1)

indexes[0]
```

```python
sns.scatterplot(x = train_df_preprocessed['TotalBsmtSF'].drop(labels=indexes[0]), y = train_df['SalePrice'])
```

```python
train_df_preprocessed = train_df_preprocessed.drop(labels=indexes[0], axis=0)
```

```python
lof = LocalOutlierFactor(novelty=False)
preds = lof.fit_predict(np.array(train_df_preprocessed["SalePrice"]).reshape(-1, 1))

indexes = np.where(preds==-1)

indexes[0]
```

```python
train_df_preprocessed = train_df_preprocessed.drop(labels=indexes[0], axis=0)
```

```python
from xgboost import XGBRegressor

model_xgb = XGBRegressor()
model_xgb.fit(train_df_preprocessed.drop("SalePrice", axis=1), train_df_preprocessed["SalePrice"])
```

```python
from catboost import CatBoostRegressor

model_ctb = CatBoostRegressor()

model_ctb.fit(train_df_preprocessed.drop("SalePrice", axis=1), train_df_preprocessed["SalePrice"])
```

```python
submission = pd.read_csv('sample_submission.csv')
submission["SalePrice"] = model_ctb.predict(test_df_preprocessed)
submission.to_csv("final_submission.csv", index=False)
```
